    인덱스  매장이름  매장주소   위도  경도

```python
import requests 
from bs4 import BeautifulSoup 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import numpy as np
import pandas as pd
import time
import folium
```

```python
url = "https://banapresso.com/store"

driver = webdriver.Chrome('D:/Coding/Classroom/python/Day2/chromedriver.exe') 
driver.get(url)

request = requests.get(url)
print(request) 
```

```python
total_stores = []
is_end = False
next_page = False
page = 1
index = 0

while not is_end :
    time.sleep(0.5)

    if page != 6 and page != 1:
        driver.find_element('xpath', '//*[@id="contents"]/article/div/section[1]/div/div[1]/div[3]/ul/li[{}]/a'.format(page)).click()
    if page == 4 and next_page == True:
        is_end = True
    elif page == 6 :
        driver.find_element('xpath', '//*[@id="contents"]/article/div/section[1]/div/div[1]/div[3]/span/a').click()
        next_page = True
        page = 1

    html = driver.page_source
    soup = BeautifulSoup(html)

    shop_list = soup.find('div', {'class' : 'store_shop_list'})

    store_name = shop_list.select('.store_name_map>b')
    store_location = shop_list.select('.store_name_map>i')
    store_address = shop_list.select('.store_name_map>span')
    print(len(store_name), len(store_location), len(store_address))
    for i in range(len(shop_list.select("ul>li"))):
        if store_name[i].text.strip() and store_location[i].text.strip() and store_address[i].text.strip():
            store = [index, store_name[i].text.strip(), store_location[i].text.strip(), store_address[i].text.strip()]
            total_stores.append(store)
            index += 1
    page += 1
```

```python
print(total_stores)
```

```python
df = pd.DataFrame(total_stores)
df.columns = ['index', '매장이름', '매장위치', '매장주소']
print(df)
```

```python
add  = df['매장주소']
add
add.to_csv("./address.csv", index=False, encoding="ms949")
```

```python
pos = pd.read_csv("./address2.csv", encoding="ms949")   
pos
```

```python
bana_df = pd.merge(df, pos, on='매장주소', how='inner')
```

```python
bana_df
```

```python
data = bana_df
bana_map = folium.Map(location = [data['위도'].mean(), data['경도'].mean()], zoom_start=12)

for i in data.index:
  bana_name = data.loc[i, "매장이름"] + ' - ' + data.loc[i, '매장주소'] 
  popup = folium.Popup(bana_name, max_width=500)
  folium.Marker(location=[data.loc[i, '위도'], data.loc[i, '경도']], popup=popup).add_to(bana_map)
```
